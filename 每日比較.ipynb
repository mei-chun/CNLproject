# Daily資料比較

## 比較當天EC資料： 拋出 vs DB

```python
import pandas as pd 


def compare_data(date, DBdata_af, Dailydata_af, DBdata_navy, Dailydata_navy):
    ## 空軍比較
    ECDBtest_af = pd.read_csv(DBdata_af)
    ECdaily_af = pd.read_csv(Dailydata_af)
    print('比對', date,'DB與Daily資料：\n')
    print('空軍DB資料筆數：', ECDBtest_af.shape[0], '，去重複筆數：', ECDBtest_af.drop_duplicates().shape[0])
    print('空軍Daily資料筆數：', ECdaily_af.shape[0], '，去重複筆數：', ECdaily_af.drop_duplicates().shape[0])
    InterSectEC_af = set(ECDBtest_af['訂單編號']) & set(ECdaily_af['訂單編號'])
    #print('空軍訂單差異：', set(ECDBtest_af['訂單編號'])-InterSectEC_af, set(ECdaily_af['訂單編號'])-InterSectEC_af)
    if ECDBtest_af.drop_duplicates().shape[0] > ECdaily_af.drop_duplicates().shape[0]:
        print('空軍DB筆數較多，相差訂單：', set(ECDBtest_af['訂單編號']) - InterSectEC_af)
        display(ECDBtest_af[ECDBtest_af['訂單編號'].isin(set(ECDBtest_af['訂單編號']) - InterSectEC_af)][['訂單編號', '會員編號', '商品上架名稱', '品項狀態']])
        
    elif ECDBtest_af.drop_duplicates().shape[0] < ECdaily_af.drop_duplicates().shape[0]:
        print('空軍DB筆數較少，相差訂單：', set(ECdaily_af['訂單編號']) - InterSectEC_af)
        display(ECdaily_af[ECdaily_af['訂單編號'].isin(set(ECdaily_af['訂單編號']) - InterSectEC_af)][['訂單編號', '會員編號', '商品上架名稱', '品項狀態']])
        
    else:
        print('空軍DB與Daily筆數相同。')
        
    
    merge_data_af = pd.merge(ECDBtest_af, ECdaily_af,on=['訂單編號', '物料編碼', '訂單項次', '單件項次'], how="outer")
    if len(set(merge_data_af[-(merge_data_af['品項狀態_x']==merge_data_af['品項狀態_y'])]['訂單編號'])) != 0:
        print('空軍品項狀態不同資料：', set(merge_data_af[-(merge_data_af['品項狀態_x']==merge_data_af['品項狀態_y'])]['訂單編號']))
        display(merge_data_af[-(merge_data_af['品項狀態_x']==merge_data_af['品項狀態_y'])][['訂單編號', '會員編號_x', '商品上架名稱_x','訂單項次', '單件項次', '品項狀態_x', '品項狀態_y']])
    else:
        print('空軍品項狀態皆相同。\n')
    
    ## 海軍比較
    ECDBtest_navy = pd.read_csv(DBdata_navy)
    ECdaily_navy = pd.read_csv(Dailydata_navy)
    
    print('海軍DB資料筆數：', ECDBtest_navy.shape[0], '，去重複筆數：', ECDBtest_navy.drop_duplicates().shape[0])
    print('海軍Daily資料筆數：', ECdaily_navy.shape[0], '，去重複筆數：', ECdaily_navy.drop_duplicates().shape[0])
    InterSectEC_navy = set(ECDBtest_navy['訂單編號']) & set(ECdaily_navy['訂單編號'])
    #print('海軍訂單差異：', set(ECDBtest_navy['訂單編號'])-InterSectEC_navy, set(ECdaily_navy['訂單編號'])-InterSectEC_navy)
    if ECDBtest_navy.drop_duplicates().shape[0] > ECdaily_navy.drop_duplicates().shape[0]:
        print('海軍DB筆數較多，相差訂單：', set(ECDBtest_navy['訂單編號']) - InterSectEC_navy)
        display(ECDBtest_navy[ECDBtest_navy['訂單編號'].isin(set(ECDBtest_navy['訂單編號']) - InterSectEC_navy)][['訂單編號', '會員編號', '商品上架名稱', '品項狀態']])
        
    elif ECDBtest_navy.drop_duplicates().shape[0] < ECdaily_navy.drop_duplicates().shape[0]:
        print('海軍DB筆數較少，相差訂單：', set(ECdaily_navy['訂單編號']) - InterSectEC_navy)
        display(ECdaily_navy[ECdaily_navy['訂單編號'].isin(set(ECdaily_navy['訂單編號']) - InterSectEC_navy)][['訂單編號', '會員編號', '商品上架名稱', '品項狀態']])
        
    else:
        print('海軍DB與Daily筆數相同。')
        
       
    merge_data_navy = pd.merge(ECDBtest_navy, ECdaily_navy,on=['訂單編號', '物料編碼', '訂單項次', '單件項次'], how="outer")
    if len(set(merge_data_navy[-(merge_data_navy['品項狀態_x']==merge_data_navy['品項狀態_y'])]['訂單編號'])) != 0:
        print('海軍品項狀態不同資料：', set(merge_data_navy[-(merge_data_navy['品項狀態_x']==merge_data_navy['品項狀態_y'])]['訂單編號'])) 
        display(merge_data_navy[-(merge_data_navy['品項狀態_x']==merge_data_navy['品項狀態_y'])][['訂單編號', '會員編號_x', '商品上架名稱_x','訂單項次', '單件項次', '品項狀態_x', '品項狀態_y']])
    else:
        print('海軍品項狀態皆相同。')
    
```
